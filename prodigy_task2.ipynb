# Performing EDA  on data 

##  1. age distribution and then plotting the max age 

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20,kde=True)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of AGE ')
plt.show()
```

```python
max_age = data['Age'].max()
```

```python
plt.axvline(x=max_age, color='green', linestyle='dashed', linewidth=2, label=f'Max Age: {max_age:.1f}')
plt.legend()
```

## 2. male - female ratio survived

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Sex', palette='Set1')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Count of Male and Female Passengers')
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.countplot(x="Sex", hue="Survived", data=data, palette=["skyblue", "pink"])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Count of Survivors by Sex')
plt.show()
```

## 3. Port of Embarkation and Ticket_class

```python
#C = Cherbourg, Q = Queenstown, S = Southampton
plt.figure(figsize=(16, 9))
sns.countplot(x="Embarked",hue='Pclass', data=data, palette=["skyblue", "pink"])
plt.xlabel('Port of Embarkation')
plt.ylabel('Ticket_class')
plt.title('Port of Embarkation and Ticket_class')
plt.show()

```

```python
max_ticketclass = data['Pclass'].max()
```

```python
plt.axvline(x=max_ticketclass, color='green', linestyle='dashed', linewidth=2, label=f'Max ticket_class: {max_age:.1f}')
plt.legend()
```

## 

## 4. Scatter plot between 'Age' and 'Fare'

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Age', y='Fare', hue='Survived', alpha=0.6)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare (Survived)')
plt.legend(title='Survived', loc='upper right', labels=['Not Survived', 'Survived'])
plt.show()
```

## 5. Fare by Sex and Survival Status

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="Sex", y="Fare", hue="Survived", data=data, palette="pastel", split=True)
plt.xlabel('Sex')
plt.ylabel('Fare')
plt.title('Violin Plot of Fare by Sex and Survival Status')
plt.legend(title='Survived', loc='upper right', labels=['Not Survived', 'Survived'])
plt.show()
```

## 6. distribution of age and fare 

```python
plt.figure(figsize=(8, 6))
sns.jointplot(x="Age", y="Fare", data=data, kind="scatter", color="purple")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title(' Age and Fare')
plt.show()
```
